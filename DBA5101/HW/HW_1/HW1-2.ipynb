# Answer 1: Linear Regression

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
# Importing data
df = pd.read_csv('homeworkI-OLS.csv')
df.head(10)
```

The dataset consists of two columns:

- **Advertising Spend (X) in $1,000s**: Represents the amount spent on advertising
  
- **Sales Revenue (Y) in $1,000s**: Represents the corresponding sales revenue

```python
# Define the independent variable (X) and dependent variable (Y)
X = df['Advertising Spend (X) in $1,000s']
Y = df['Sales Revenue (Y) in $1,000s']

# Add a constant to the independent variable (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(Y, X).fit()

# Output the summary of the regression
model_summary = model.summary()
model_summary
```


### 1. **Estimated Regression Line**

The regression equation based on the results is:

$$\text{Sales Revenue (Y)} = 7.2295 + 1.9399 \times \text{ Advertising Spend (X)} + \epsilon $$

```python
# Fit a simple linear regression model
residuals = model.resid
fitted_values = model.fittedvalues

X = df['Advertising Spend (X) in $1,000s']
Y = df['Sales Revenue (Y) in $1,000s']

plt.scatter(X, Y, color='blue', alpha=0.6)
plt.plot(X, fitted_values, color='red', linewidth=2)
plt.title('Linearity Assumption: Scatter Plot with Regression Line')
plt.xlabel('Advertising Spend (X) in $1,000s')
plt.ylabel('Sales Revenue (Y) in $1,000s')
plt.grid(True)
```

### 2. **Interpretation of the Regression Results**


- **Intercept ($\beta_0 = 7.2295$)**: When advertisement spending is `$0`, the expected sales revenue is `$7,229.5` (in $1,000s).

  
- **Slope ($\beta_1 = 1.9399$)**: For every additional `$1,000` spent on advertising, the sales revenue is expected to increase by `$1,939.9` (in $1,000s).

- **Significance of the Slope Coefficient**:
  - The p-value associated with the slope coefficient ($\beta_1\$) is extremely small (essentially 0), indicating that advertising spending has a statistically significant impact on sales revenue.


### Homoskedasticity Test

```python
plt.scatter(X, residuals, color='green', alpha=0.6)
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Homoskedasticity Assumption: Residuals vs. Independent Variable')
plt.xlabel('Advertising Spend (X) in $1,000s (X)')
plt.ylabel('Residuals')
plt.grid(True)
```

### The plot shows a random scatter with a consistent spread of residuals across different levels of advertising spending, this suggests that the model satisfies the assumption of homoskedasticity.

### Normality Test

```python
plt.hist(residuals, color='orange', alpha=0.6)
plt.title('Normality Assumption: Histogram of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
```

### The histogram appears and bell-shaped, this suggests that the residuals are approximately normally distributed, which is an assumption of ordinary least squares regression.

# Answer 2: Market Structure and Monopoly

## Monopolist Problem

### Given Information:
- **Marginal Cost (MC):** $4
  
- **Fixed Cost (FC):** $48
  
- **Demand Curve:** $( Q = 12 - \frac{P}{2}$\), where $P$ is the price and $Q$ is the demand quantity.

### 1. Drawing the Demand Curve and Marginal Cost Curve
First, express the demand curve in terms of \( P \):


$Q = 12 - \frac{P}{2} \implies P = 24 - 2Q$


So, the demand curve is:


$ P = 24 - 2Q $


The marginal cost (MC) is constant at $4.

---


### 2. Finding the Marginal Revenue (MR)
The total revenue (TR) is given by: $Price \times Quantity$


$ TR = P \times Q = (24 - 2Q) \times Q = 24Q - 2Q^2 $


The marginal revenue (MR) is the derivative of TR with respect to Q:


$ MR = \frac{d(TR)}{dQ} = \frac{d(24Q - 2Q^2)}{dQ} = 24 - 4Q $


So, the marginal revenue function is:


$ MR = 24 - 4Q $

---

### 3. Profit-Maximizing Price and Quantity
To maximize profit, the monopolist will produce where $ MR = MC $.

Setting $ MR = MC $:


$ 24 - 4Q = 4 $


Solving for $ Q $:


$ Q = \frac{20}{4} = 5 $


Substitute $ Q = 5 $ into the demand curve $ P = 24 - 2Q $to find the price $ P $:


$ P = 24 - 2(5) = 14 $

So, the profit-maximizing price is $ P = 14 $ and the profit-maximizing quantity is $ Q = 5 $

```python
import numpy as np
import matplotlib.pyplot as plt

# Demand curve: P = 24 - 2Q
Q = np.linspace(0, 12, 100)
P = 24 - 2 * Q

# Marginal Revenue curve: MR = 24 - 4Q
MR = 24 - 4 * Q

# Marginal Cost curve: MC = 4 (constant)
MC = np.full_like(Q, 4)

# Plotting the curves
plt.figure(figsize=(10, 6))
plt.plot(Q, P, label="Demand Curve: P = 24 - 2Q", color='blue')
plt.plot(Q, MR, label="Marginal Revenue: MR = 24 - 4Q", color='green')
plt.plot(Q, MC, label="Marginal Cost: MC = 4", color='black', linestyle='--')

# Marking the profit-maximizing point
Q_optimal = 5
P_optimal = 14
plt.scatter(Q_optimal, P_optimal, color='red')
plt.text(Q_optimal + 1.7, P_optimal + 1.7, f"Profit-maximizing point (Q={Q_optimal}, P={P_optimal})", 
         fontsize=10, ha='center')

# Labels and legend
plt.xlabel('Quantity (Q)')
plt.ylabel('Price / Cost (in $)')
plt.title('Demand, Marginal Revenue, and Marginal Cost Curves')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```


### 4. Monopolist’s Profits at the Profit-Maximizing Price

The profit-maximizing price $P$ and quantity $Q$ have already been calculated:

- **Profit-maximizing price:** \( P = 14 \)
- **Profit-maximizing quantity:** \( Q = 5 \)

**Total Revenue (TR):**

$TR = P \times Q = 14 \times 5 = 70$


**Total Cost (TC):**

Total cost is the sum of the total variable cost (TVC) and the fixed cost (FC):

- **Total Variable Cost (TVC):** Given by $MC \times Q = 4 \times 5 = 20$
- **Fixed Cost (FC):** $48$

So, the total cost $TC$ is:


$TC = TVC + FC = 20 + 48 = 68$


**Profit:**


$\text{Profit} = TR - TC = 70 - 68 = 2$


Thus, the monopolist's profit at the profit-maximizing price is **$2**.

---

### 5. Finding the Price and Quantity Where the Monopolist Just Covers Its Costs

To find the price and quantity where the monopolist just covers its costs (i.e., makes zero profit), we need to find the price \( P \) and quantity \( Q \) where **Total Revenue = Total Cost**.

**Step 1: Find the Average Cost (AC) Function**

The total cost (TC) function is:


$TC = TVC + FC = 4Q + 48$


The average cost (AC) is:


$AC = \frac{TC}{Q} = \frac{4Q + 48}{Q} = 4 + \frac{48}{Q}$


**Step 2: Set Price Equal to Average Cost**

From the demand curve, $P = 24 - 2Q $. At zero profit, $P = AC $:


$24 - 2Q = 4 + \frac{48}{Q} $


**Step 3: Solve for Quantity \( Q \)**

Rearrange the equation to:


$20 - 2Q = \frac{48}{Q}$


Multiply through by \( Q \) to get rid of the fraction:


$20Q - 2Q^2 = 48$


Rearrange into a standard quadratic equation form:


$2Q^2 - 20Q + 48 = 0$


**Step 4: Solve the Quadratic Equation**

We solve this quadratic equation using the quadratic formula:


$Q = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$


Where \( a = 2 \), \( b = -20 \), and \( c = 48 \):



The solutions to the quadratic equation are $Q = 6$ and $Q = 4$.

**Step 5: Determine the Corresponding Price**

Now, substitute these quantities back into the demand curve to find the corresponding prices.

For $Q = 6 $:


$P = 24 - 2(6) = 12$


For $Q = 4$:


$P = 24 - 2(4) = 16$


### **Interpretation:**

- At  $Q = 6 $ and $ P = 12 $ : The monopolist covers its costs, generating zero profit.
- At $ Q = 4 $ and $ P = 16 $ : The monopolist also covers its costs, generating zero profit.

These are the two points where the monopolist just covers its costs, depending on the price and quantity combination chosen.
