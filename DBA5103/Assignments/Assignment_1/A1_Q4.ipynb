```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog

# Coefficients of the objective function (to maximize profit Z)
c = [-4, -3]  # Negative because linprog does minimization by default

# Coefficients of the inequalities (Ax <= b)
A = [[6, 4], [2, 3], [5, 2]]
b = [2400, 800, 1200]

# Bounds for x1 and x2 (non-negativity)
x_bounds = (0, None)
y_bounds = (0, None)

# Use linprog to find the optimal solution
result = linprog(c, A_ub=A, b_ub=b, bounds=[x_bounds, y_bounds], method='highs')

# Extracting results
x1_opt, x2_opt = result.x
optimal_profit = 4 * x1_opt + 3 * x2_opt

# Calculate the intersection points of the constraints for plotting
x1 = np.linspace(0, 500, 500)

# Equations for constraints
y1 = (2400 - 6 * x1) / 4
y2 = (800 - 2 * x1) / 3
y3 = (1200 - 5 * x1) / 2

# Plotting the feasible region
plt.figure(figsize=(8, 8))
plt.plot(x1, y1, label=r'$6x_1 + 4x_2 \leq 2400$')
plt.plot(x1, y2, label=r'$2x_1 + 3x_2 \leq 800$')
plt.plot(x1, y3, label=r'$5x_1 + 2x_2 \leq 1200$')

# Shading the feasible region
plt.fill_between(x1, np.minimum(np.minimum(y1, y2), y3), where=(y1 >= 0) & (y2 >= 0) & (y3 >= 0), color='grey', alpha=0.3)

# Marking the optimal point
plt.plot(x1_opt, x2_opt, 'ro', label=f'Optimal Solution (x1={x1_opt:.2f}, x2={x2_opt:.2f})')
plt.xlim((0, 500))
plt.ylim((0, 500))

plt.xlabel(r'$x_1$ (Special Risk Insurance)')
plt.ylabel(r'$x_2$ (Mortgages)')
plt.title('Feasible Region and Optimal Solution')
plt.legend()
plt.grid(True)
plt.show()

x1_opt, x2_opt, optimal_profit
```
